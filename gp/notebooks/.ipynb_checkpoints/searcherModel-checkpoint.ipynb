```python
params = { 'xlims':(-50.0, 50.0), 'ylims':(-50.0, 50.0), 'zlims':(-50.0, 50.0), 'res':0.25 }
params.update({ \
               'random_pos':True, \
               'random_motion_init': True, \
               'num_iter':100, \
               'num_humans':1, \
               'num_searchers':2, \
               'search_steps':100 \
              })
```
